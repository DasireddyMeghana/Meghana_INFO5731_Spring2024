## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):


reviews = [
    "This movie was an excellent portrayal of a historical event, with amazing acting and a gripping plot.",
    "I found the movie boring and too long. The plot was predictable and uninteresting.",
    "A masterpiece of animation and storytelling, blending humor and emotion seamlessly.",
    "The movie was okay, but it lacked originality and depth in its storyline.",
    "Terrible movie. It was a complete waste of time, with a dull storyline and poor acting."
]
```

```python
#1. Bag of Words (BoW)
from sklearn.feature_extraction.text import CountVectorizer

# Initialize a CountVectorizer object to convert text data into a matrix of token counts
vectorizer = CountVectorizer()

# This converts the text into a sparse matrix of word counts.
X_bow = vectorizer.fit_transform(reviews)

print("Bag of Words Feature Matrix:\n", X_bow.toarray())
print("Feature Names:\n", vectorizer.get_feature_names_out())
```

```python
#2. TF-IDF (Term Frequency-Inverse Document Frequency)

from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize a TfidfVectorizer object to transform the text data into TF-IDF features.
tfidf_vectorizer = TfidfVectorizer()

X_tfidf = tfidf_vectorizer.fit_transform(reviews)

# Printing the resulting TF-IDF feature matrix as an array. The matrix representation shows the TF-IDF scores,
print("TF-IDF Feature Matrix:\n", X_tfidf.toarray())

# Printing the feature names associated with the columns in the TF-IDF matrix.
print("Feature Names:\n", tfidf_vectorizer.get_feature_names_out())
```

```python
#3. Sentiment Lexicons
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()

# Example of extracting sentiment scores for the first review
print("Sentiment scores for the first review:\n", sid.polarity_scores(reviews[0]))
```

```python
#4. Part-of-Speech (POS) Tagging
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

# Function to extract POS tags for each review
def extract_pos_tags(texts):
    pos_tags = []
    for text in texts:
        tokens = word_tokenize(text)  # Tokenize the text
        tags = pos_tag(tokens)  # Get POS tags for each token
        pos_tags.append(tags)
    return pos_tags

# Extract POS tags for the reviews
pos_tags_reviews = extract_pos_tags(reviews)

# Example: Print POS tags for the first review
print("POS Tags for the first review:\n", pos_tags_reviews[0])
```

```python
#5. N-grams
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))  # Considering unigrams and bigrams
X_ngrams = ngram_vectorizer.fit_transform(reviews)

print("N-gram Feature Matrix:\n", X_ngrams.toarray())
print("Feature Names:\n", ngram_vectorizer.get_feature_names_out())
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import chi2
import numpy as np

# Movie reviews dataset
# reviews = [
#     "This movie was an excellent portrayal of a historical event, with amazing acting and a gripping plot.",
#     "I found the movie boring and too long. The plot was predictable and uninteresting.",
#     "A masterpiece of animation and storytelling, blending humor and emotion seamlessly.",
#     "The movie was okay, but it lacked originality and depth in its storyline.",
#     "Terrible movie. It was a complete waste of time, with a dull storyline and poor acting."
# ]

# Assuming labels for each review
review_labels = ['positive', 'negative', 'positive', 'neutral', 'negative']

# Convert labels to numerical values
labels = ['positive', 'negative', 'neutral']
label_map = {label: idx for idx, label in enumerate(labels)}
numerical_labels = [label_map[label] for label in review_labels]

# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(reviews)

# Chi-squared test for feature selection
chi2_scores, _ = chi2(tfidf_features, numerical_labels)

# Create a dictionary to associate feature names with chi-squared scores
feature_scores = {feature_name: score for feature_name, score in zip(tfidf_vectorizer.get_feature_names_out(), chi2_scores)}

# Sort features by their chi-squared scores in descending order
sorted_features = sorted(feature_scores.items(), key=lambda x: x[1], reverse=True)

# Print the top N features and their scores
top_n = 15  # Adjust this to display the desired number of top features
print("Top features based on chi-squared scores:")
for feature, score in sorted_features[:top_n]:
    print(f"Feature: {feature}, Chi-squared Score: {score:.2f}")


```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):

from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained model tokenizer (vocabulary) and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to get BERT embeddings for a text
def get_bert_embedding(text):
    # Encode text
    encoded_input = tokenizer(text, return_tensors='pt', truncation=True, max_length=512)
    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input)
    # Pool the outputs into a single mean vector
    embeddings = model_output.last_hidden_state.mean(dim=1)
    return embeddings

# reviews = [
#     "This movie was an excellent portrayal of a historical event, with amazing acting and a gripping plot.",
#     "I found the movie boring and too long. The plot was predictable and uninteresting.",
#     "A masterpiece of animation and storytelling, blending humor and emotion seamlessly.",
#     "The movie was okay, but it lacked originality and depth in its storyline.",
#     "Terrible movie. It was a complete waste of time, with a dull storyline and poor acting."
# ]

query = "An engaging story with deep characters and outstanding performances."

# Get embeddings for the query and the documents
query_embedding = get_bert_embedding(query)
doc_embeddings = torch.stack([get_bert_embedding(review) for review in reviews])

# Convert embeddings to 2D arrays if necessary
query_embedding_2d = query_embedding.numpy()
doc_embeddings_2d = doc_embeddings.numpy().squeeze(axis=1)

# Calculate cosine similarities between query and documents
cos_similarities = cosine_similarity(query_embedding_2d, doc_embeddings_2d)[0]

# Rank documents based on similarity
sorted_doc_indices = cos_similarities.argsort()[::-1]

print("Ranking of documents based on similarity to the query:")
for rank, index in enumerate(sorted_doc_indices):
    print(f"{rank+1}. Document {index+1} (Similarity: {cos_similarities[index]:.4f})")
    print(f"   Text: {reviews[index]}")



```
